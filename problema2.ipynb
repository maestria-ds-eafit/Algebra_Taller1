```python
import pandas as pd
```

```python
file_path = 'general_science.txt'  # Replace 'your_file.txt' with the path to your text file
try:
    with open(file_path, 'r') as file:
        # Read the content of the file and split it into words
        words = [word.strip(",") for word in file.read().split()]
except FileNotFoundError:
    print(f"The file '{file_path}' was not found.")
    words = []

# Now 'words' contains a list of all the words in the text file
print(words)
```

```python
def levenshtein_distance(s1:str, s2:str):
    # Create a matrix to store the distances between substrings
    matrix = [[0] * (len(s2) + 1) for _ in range(len(s1) + 1)]

    # Initialize the first row and column
    for i in range(len(s1) + 1):
        matrix[i][0] = i
    for j in range(len(s2) + 1):
        matrix[0][j] = j

    # Fill in the matrix using dynamic programming
    for i in range(1, len(s1) + 1):
        for j in range(1, len(s2) + 1):
            cost = 0 if s1[i - 1] == s2[j - 1] else 1
            matrix[i][j] = min(
                matrix[i - 1][j] + 1,       # Deletion
                matrix[i][j - 1] + 1,       # Insertion
                matrix[i - 1][j - 1] + cost  # Substitution
            )

    # The final value in the matrix represents the Levenshtein distance
    return matrix[len(s1)][len(s2)]
```

```python
def calculate_distances(word, word_list):
    distances = {}
    for w in word_list:
        distance = levenshtein_distance(word, w)
        distances[w] = distance
    return distances
```

```python
word = str(input("una palabra: "))
distances = calculate_distances(word, words)
auto = {}
word_len = len(word)
for w, distance in distances.items():
    if w [:word_len] == word:
        auto[w]=distance

print(auto)
```

```python
min_key = min(auto, key=distances.get)
print(word,min_key)
```
